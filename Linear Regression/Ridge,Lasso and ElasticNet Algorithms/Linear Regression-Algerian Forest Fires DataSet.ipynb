### Data Cleaning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
df.head()   
```

```python
# Data Cleaning
df[df.isnull().any(axis=1)]
```

- The Dataset is converted into two regions from 122th index,we can make a column wrt regions
- Bejaia Region Dataset
- Sidi-Bel Abbes Region Dataset

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df['Region'] = df['Region'].astype('int64')	
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns  
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
df[["month","day","year","Temperature","RH","Ws"]]=df[["month","day","year","Temperature","RH","Ws"]].astype(int)
```

```python
df.info()
```

```python
df["DC"]=df["DC"].str.replace("14.6 9","14.69")
```

```python
df["DC"].unique()
```

```python
df["FWI"]=df["FWI"].str.replace("fire   ","0")
```

```python
for i in df.columns:
   if df[i].dtype == 'object':
        if i!="Classes":
          print(i)
          df[i]=df[i].astype(float)
```

```python
df.to_csv('Algerian_forest_fires_dataset_Cleaned.csv',index=False)
```

### Exploratory Data Analysis

```python
df_copy=df.copy()
```

```python
df_copy.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df_copy.head()
```

```python
# Target Ordinal Encoding
df["Classes"].unique()  
```

```python
df_copy["Classes"].value_counts()
```

```python
df_copy["Classes"]=df_copy["Classes"].map({"fire   ":1,'fire':1,"fire ":1,'not fire':0,"not fire   ":0,"not fire ":0,"not fire     ":0,"not fire    ":0})
```

```python
df["Classes"].value_counts()
```

```python
# Density Plot
plt.figure(figsize=(10,10))
plt.style.use('ggplot')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
df_copy["Classes"].value_counts(normalize=True)*100
plt.figure(figsize=(5,5))
plt.pie(df_copy["Classes"].value_counts(normalize=True)*100,labels=["Fire","Not Fire"],autopct='%1.1f%%', startangle=140)
plt.title("Classes Distribution")
```

```python
# Corelation 
plt.figure(figsize=(10,10))
sns.heatmap(df_copy.corr(),annot=True,cmap='coolwarm',linewidths=0.5)
```

```python
# Boxplot of FWI in horizontal direction
sns.boxplot(df["FWI"],color='red',orient='h')
```

```python
# Monthly Fire Analysis
df["Classes"]=df["Classes"].map({"fire   ":1,'fire':1,"fire ":1,'not fire':0,"not fire   ":0,"not fire ":0,"not fire     ":0,"not fire    ":0})
dftemp=df[df["Region"]==1]
plt.subplots(figsize=(10,10))
plt.xlabel("Month",weight="bold")
plt.ylabel("Fire",weight="bold")
plt.title("Monthly Fire Analysis of Sidi-Bel Abbes Region")
sns.set_style("whitegrid")
sns.countplot(x="month",data=df,hue="Classes",palette="viridis")


```

```python
# Monthly Fire Analysis
dftemp=df[df["Region"]==0]
plt.subplots(figsize=(10,10))
plt.xlabel("Month",weight="bold")
plt.ylabel("Fire",weight="bold")
plt.title("Monthly Fire Analysis of bijaia Region")
sns.set_style("whitegrid")
sns.countplot(x="month",data=df,hue="Classes",palette="viridis")


```
